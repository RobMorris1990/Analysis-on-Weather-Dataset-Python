### Adding Fahrenheit and Kelvin columns to the dataframe

```python
def conv_c_to_f(Temp_C):
    return (Temp_C * (9/5)) + 32
```

```python
data["Fahrenheit"] = data["Temp_C"].apply(conv_c_to_f)
```

```python
data
```

```python
def conv_c_to_k(Temp_C):
    return Temp_C + 273.15
```

```python
data["Kelvin"] = data["Temp_C"].apply(conv_c_to_k)
```

```python
data
```
